```python
#importing needed libraries
import numpy as np
import pandas as pd

#Load dataset
transactions = pd.read_csv("datasets/nigerian_market_transactions_assignment.csv")
prices = pd.read_csv("datasets/nigerian_market_prices_assignment.csv")
```

```python
#Derive price from the transactions revenue
transactions["price"] = transactions["revenue"]/transactions["quantity"]
```

```python
transactions.head()
```

```python
transactions.isnull().sum()
```

```python
prices.isnull().sum()
```

```python
#Handling missing values from price
#prices["price"] = prices["price"].str.replace('₦', '').str.replace(',', '')

# Step 2: Convert to numeric, turning any invalid values (including NaN) to NaN
prices["price"] = pd.to_numeric(prices["price"])
# Step 3: Now fill the NaN values with the median
prices["price"] = prices["price"].fillna(prices["price"].median())

prices.head()
```

```python
prices.isnull().sum()
```

```python
#Converting the date columns to actual datetime format
transactions["transaction_date"] = pd.to_datetime(transactions["transaction_date"], errors="coerce")
```

```python
category_summary = transactions.groupby("product").agg({'revenue' : ['mean']})
```

```python
print(category_summary)
```

```python
category_summary = transactions.groupby("product").agg({
    "price": ['mean'],
    "quantity": "sum",
    "product": "count"
})
```

```python
print(category_summary)
```

```python
#Merging dataset
merged_data = pd.merge(transactions, prices, on = "product", how = "inner")
```
